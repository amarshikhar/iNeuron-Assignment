Question 1: -
Write a program that takes a string as input, and counts the frequency of each word in the string, there might
be repeated characters in the string. Your task is to find the highest frequency and returns the length of the
highest-frequency word.

Note - You have to write at least 2 additional test cases in which your program will run successfully and provide
an explanation for the same.
Example input - string = “write write write all the number from from from 1 to 100”
Example output - 5
Explanation - From the given string we can note that the most frequent words are “write” and “from” and
the maximum value of both the values is “write” and its corresponding length is 5

```python
def highest_frequency_word_length(string):
    # Split the string into words
    words = string.split()

    # Count the frequency of each word
    word_count = {}
    for word in words:
        word_count[word] = word_count.get(word, 0) + 1

    # Find the highest frequency
    highest_frequency = max(word_count.values())

    # Find the words with the highest frequency
    highest_frequency_words = [word for word, count in word_count.items() if count == highest_frequency]

    # Find the length of the highest-frequency words
    highest_frequency_word_lengths = [len(word) for word in highest_frequency_words]

    explanation = "From the given string, we can note that the most frequent words are: "
    explanation += ", ".join(highest_frequency_words)
    explanation += " and the maximum value of both the frequencies is " + str(highest_frequency)
    explanation += ". The lengths of the highest-frequency words, " + str(highest_frequency_words) + ", are " + str(highest_frequency_word_lengths)

    return highest_frequency_word_lengths, explanation


string = "write write write all all all the number from from from 1 to 100"
result, explanation = highest_frequency_word_length(string)
print(result)
print(explanation)
```

Question 2.
Consider a string to be valid if all characters of the string appear the same number of times. It is also valid if
he can remove just one character at the index in the string, and the remaining characters will occur the same
number of times. Given a string, determine if it is valid. If so, return YES , otherwise return NO .

Note - You have to write at least 2 additional test cases in which your program will run successfully and provide
an explanation for the same.
Example input 1 - s = “abc”. This is a valid string because frequencies are { “a”: 1, “b”: 1, “c”: 1 }
Example output 1- YES
Example input 2 - s “abcc”. This string is not valid as we can remove only 1 occurrence of “c”. That leaves
character frequencies of { “a”: 1, “b”: 1 , “c”: 2 }
Example output 2 - NO

```python
def isValidString(s):
    freq = {}
    for char in s:
        freq[char] = freq.get(char, 0) + 1

    counts = list(freq.values())
    unique_counts = set(counts)

    if len(unique_counts) == 1:
        return "YES"  # All characters appear the same number of times

    if len(unique_counts) == 2:
        count1, count2 = unique_counts
        if counts.count(count1) == 1 and (count1 == 1 or count1 - 1 == count2):
            return "YES"  # Removing one character makes all characters appear the same number of times

    return "NO"

# Example usage:
print(isValidString("abc"))  # Output: YES
print(isValidString("abcc"))  # Output: NO
```

Question 8 -
Using the data from Question 5, write code the analyze the data and answer the following questions Note -
1. Draw plots to demonstrate the analysis for the following questions and better visualizations
2. Write code comments wherever required for code understanding

```python
import requests
import matplotlib.pyplot as plt

# Make the HTTP request to retrieve the data
response = requests.get("http://api.tvmaze.com/singlesearch/shows?q=westworld&embed=episodes")
data = response.json()

# Initialize empty lists to store ratings and episode information
season_ratings = []
episode_ratings = {}
episode_names_above_8 = {}
episode_names_before_may_2019 = {}
episode_names_highest_rating = {}
episode_names_lowest_rating = {}
episode_summary_most_popular = {}

# Iterate over each episode to extract ratings and other information
for episode in data["_embedded"]["episodes"]:
    season_number = episode["season"]
    rating = episode["rating"]["average"]
    episode_name = episode["name"]
    air_date = episode["airdate"]

    # Add the rating to the corresponding season's list
    while len(season_ratings) < season_number:
        season_ratings.append([])
    season_ratings[season_number - 1].append(rating)

    # Check if the average rating is above 8 and store the episode name
    if rating > 8:
        if season_number not in episode_names_above_8:
            episode_names_above_8[season_number] = []
        episode_names_above_8[season_number].append(episode_name)

    # Check if the air date is before May 2019 and store the episode name
    if air_date < "2019-05-01":
        if season_number not in episode_names_before_may_2019:
            episode_names_before_may_2019[season_number] = []
        episode_names_before_may_2019[season_number].append(episode_name)

    # Store the episode name with the highest rating for each season
    if season_number not in episode_names_highest_rating or rating > episode_names_highest_rating[season_number][1]:
        episode_names_highest_rating[season_number] = (episode_name, rating)

    # Store the episode name with the lowest rating for each season
    if season_number not in episode_names_lowest_rating or rating < episode_names_lowest_rating[season_number][1]:
        episode_names_lowest_rating[season_number] = (episode_name, rating)

    # Store the summary for the most popular episode in each season
    if season_number not in episode_summary_most_popular or rating > episode_summary_most_popular[season_number][1]:
        episode_summary_most_popular[season_number] = (episode_name, rating, episode["summary"])

# Plotting the ratings for each season
plt.figure(figsize=(10, 6))
for season in range(len(season_ratings)):
    plt.plot(range(1, len(season_ratings[season]) + 1), season_ratings[season], label=f"Season {season + 1}")
plt.xlabel("Episode Number")
plt.ylabel("Average Rating")
plt.title("Ratings Comparison for All Seasons")
plt.legend()
plt.show()

# Print episode names with average rating above 8 for each season
for season, episode_names in episode_names_above_8.items():
    print(f"Season {season}:")
    for name in episode_names:
        print(f"- {name}")
    print()

# Print episode names that aired before May 2019 for each season
for season, episode_names in episode_names_before_may_2019.items():
    print(f"Season {season}:")
    for name in episode_names:
        print(f"- {name}")
    print()

# Print episode name with the highest and lowest rating for each season
for season, (highest_name, highest_rating) in episode_names_highest_rating.items():
    print(f"Highest rated episode of Season {season}: {highest_name} (Rating: {highest_rating})")

for season, (lowest_name, lowest_rating) in episode_names_lowest_rating.items():
    print(f"Lowest rated episode of Season {season}: {lowest_name} (Rating: {lowest_rating})")

# Print the summary for the most popular episode in each season
for season, (popular_name, popular_rating, popular_summary) in episode_summary_most_popular.items():
    print(f"Most popular episode of Season {season}: {popular_name} (Rating: {popular_rating})")
    print(f"Summary: {popular_summary}")
    print()

```

Question 10 -
Write a program to count the number of verbs, nouns, pronouns, and adjectives in a given particular phrase or
paragraph, and return their respective count as a dictionary.
Note -
1. Write code comments wherever required for code
2. You have to write at least 2 additional test cases in which your program will run successfully and provide
an explanation for the same.

```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.tag import pos_tag
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
def count_pos_tags(text):
    # Tokenize the text into words
    words = word_tokenize(text)
    
    # Perform Part-of-Speech tagging
    pos_tags = pos_tag(words)
    
    # Initialize counts dictionary
    counts = {
        'Noun': 0,
        'Verb': 0,
        'Pronoun': 0,
        'Adjective': 0
    }
    
    # Iterate over tagged words and count each POS category
    for word, tag in pos_tags:
        if tag.startswith('N'):  # Noun
            counts['Noun'] += 1
        elif tag.startswith('V'):  # Verb
            counts['Verb'] += 1
        elif tag.startswith('P'):  # Pronoun
            counts['Pronoun'] += 1
        elif tag.startswith('J'):  # Adjective
            counts['Adjective'] += 1
    
    return counts

# Example usage
text = "The quick brown fox jumps over the lazy dog."
result = count_pos_tags(text)
print(result)
```
